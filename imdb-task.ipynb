# **Step 2: Clean and preprocess the data by removing unnecessary columns and handling missing values.**

```python
IMDB_Dataset.drop(['Poster_Link','Overview'],axis=1,inplace=True)
IMDB_Dataset
```

```python
missing_values = IMDB_Dataset.isna().sum().reset_index()
missing_values.columns = ['columns','missing_counts']
missing_values
```

So we have 3 columns which have missing values:

* Certificate
* Meta_score
* Gross

```python
IMDB_Dataset['Certificate'].value_counts()
```

```python
# As most movies are `U` rated so we are setting null/empty values as `U`
IMDB_Dataset['Certificate'].fillna('U',inplace=True)
```

```python
IMDB_Dataset['Meta_score'].value_counts()
```

```python
# Setting meta-score's null/empty value to average meta_score.
IMDB_Dataset['Meta_score'].fillna(IMDB_Dataset['Meta_score'].mean(),inplace=True)
```

```python
IMDB_Dataset['Gross'].value_counts()
```

```python
# replacing `,` with no space
IMDB_Dataset['Gross'] = IMDB_Dataset['Gross'].str.replace(',','')

# converting object to float
IMDB_Dataset['Gross'] = IMDB_Dataset['Gross'].astype(float)

# filling NaN values with Gross mean value
IMDB_Dataset['Gross'] = IMDB_Dataset['Gross'].fillna(IMDB_Dataset['Gross'].mean())

# converting float type to int
IMDB_Dataset['Gross'] = IMDB_Dataset['Gross'].astype(int)

IMDB_Dataset
```

```python
# Verifying missing values
IMDB_Dataset.info()
```

```python
# Converting Runtime from object to int
IMDB_Dataset['Runtime'] = IMDB_Dataset['Runtime'].str.replace(' min','')
IMDB_Dataset['Runtime'] = IMDB_Dataset['Runtime'].astype(int)
IMDB_Dataset['Runtime'].dtype
```

# **Step 3 : Explore the data to identify the most popular genre of movies and TV shows using pandas.**

```python


genre_counts = pd.DataFrame(IMDB_Dataset['Genre'].value_counts())
genre_counts.rename(columns={"Genre": "Number of Genre"}, inplace=True)
total_count = genre_counts["Number of Genre"].sum()
genre_counts["Genre %"] = round(genre_counts["Number of Genre"] / total_count * 100, 2)
genre_counts["Name"] = genre_counts.index

# Pie plot
fig = px.pie(genre_counts,
             values="Genre %",
             names="Name",
             title="The most popular genre of movies and TV shows",
             template="seaborn",
             width=1000,
             height=1000)
fig.update_traces(textposition="inside", textinfo="value+percent+label")
fig.show()
```

# **Step 4 : Create a new dataframe containing the count of movies/TV shows in each genre using pandas.**

```python

genre_counts = IMDB_Dataset['Genre'].value_counts().to_frame().rename(columns={'Genre': 'Count'})
genre_counts
```

# **Step 5 : Visualize the data using a bar chart to show the most popular genre of movies/TV shows using pandas and the matplotlib library.**

```python
genre_counts.head(6).plot(kind='bar', legend=False)
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Most Popular Genre of Movies/TV Shows')
plt.show()
```

# **Step 6 : Identify the director with the most top-rated movies/TV shows by creating a new dataframe containing the count of movies/TV shows directed by each director using pandas.**

```python
director_counts = IMDB_Dataset['Director'].value_counts().to_frame().rename(columns={'Director': 'Count'})
director_counts.reset_index(inplace=True)
director_counts.rename(columns={'index': 'Director'}, inplace=True)
director_counts



```

# Step 7 : Visualize the data using a bar chart to show the director with the most top-rated movies/TV shows using pandas and matplotlib. 

```python
director_counts.head(10).plot(x='Director', y='Count', kind='bar')
plt.xlabel('Director')
plt.ylabel('Count')
plt.title('Director with the Most Top-Rated Movies/TV Shows')
plt.show()
```
